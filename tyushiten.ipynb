```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 背景画像を読み込む
# 例: "background.jpg" は任意の画像ファイル名
img = mpimg.imread("../koshiten.png")

# 図の作成
fig, ax = plt.subplots(figsize=(10, 5))

# 背景画像を表示
# extentで画像を0~1920,0~1080に広げる
ax.imshow(img, extent=[0, 1920, 0, 1080])

# 軸の範囲を設定
ax.set_xlim(0, 1920)
ax.set_ylim(0, 1080)

# 中心の座標
center_x = 1920 / 2
center_y = 1080 / 2

# 中心に点をプロット
ax.plot(center_x, center_y, 'ro', markersize=30,alpha=0.5)  # 'ro'は赤い点

# タイトル（任意）
ax.set_title("中心にプロット")

# 表示
plt.show()
```
